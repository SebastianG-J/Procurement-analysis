### Group by supplier and posttype, count rows, then pivot so each posttype is its own column


```python
# Remap Montageforbrug → Salg in-place
sales_df['Posttype'] = sales_df['Posttype'].replace({'Montageforbrug': 'Salg'})

# Now rerun your three checks:

# 1. How many unique post types are there?
num_posttypes = sales_df['Posttype'].nunique()
print(f"Number of unique Posttype values: {num_posttypes}\n")

# 2. What are those unique values?
unique_posttypes = sales_df['Posttype'].unique()
print("Unique Posttype values:")
print(unique_posttypes, "\n")

# 3. How many rows belong to each post type?
posttype_counts = (
    sales_df['Posttype']
      .value_counts()          # count occurrences
      .reset_index()           # DataFrame with columns ['index','Posttype']
      .rename(columns={'index':'Posttype'})
)
from IPython.display import display
display(posttype_counts)
```

### The amount of times we sell and buy within the period 01-05-2023 -- 30-04-2025

```python
supplier_posttype_counts = (
    sales_df
      .groupby(['Leverandørnr.', 'Posttype'])
      .size()
      .unstack(fill_value=0)    # turns Posttype values into columns, missing → 0
      .rename_axis(None, axis=1)
)

supplier_posttype_counts_sorted = supplier_posttype_counts.sort_values(
    by='Salg',
    ascending=False
)

display(supplier_posttype_counts_sorted.head(20))
```
